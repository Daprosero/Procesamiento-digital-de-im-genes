**Introducción a Matplotlib**

Matplotlib es una biblioteca de visualización de datos en 2D y 3D extremadamente poderosa y flexible diseñada para Python. Es una herramienta fundamental para la ciencia y análisis de datos, ya que permite crear gráficos de alta calidad con relativa facilidad.

**¿Por qué es importante en el ecosistema de Python para visualización de datos?**

La importancia de Matplotlib en el ecosistema de Python para visualización de datos radica en su versatilidad y su capacidad para generar una amplia gama de gráficos con un alto grado de personalización. Python, como lenguaje de programación, ha ganado una gran popularidad en el campo de la ciencia de datos y el análisis de datos debido a su simplicidad, flexibilidad y a la rica variedad de bibliotecas disponibles. Matplotlib complementa esta fortaleza al proporcionar una forma poderosa de visualizar y comunicar los resultados de análisis de datos y experimentos.

**Breve historia y contexto sobre su desarrollo y popularidad**

Matplotlib fue creado por John D. Hunter en 2003 como parte de su tesis doctoral en neurobiología. La motivación detrás de su creación fue la necesidad de una herramienta que pudiera visualizar datos de manera efectiva en entornos científicos y de investigación. Desde entonces, ha experimentado un crecimiento constante y se ha convertido en la biblioteca de visualización de datos de facto en Python.

La popularidad de Matplotlib se debe a varios factores, entre ellos su facilidad de uso, su amplia gama de capacidades de visualización y su integración con otras bibliotecas populares de Python, como NumPy y Pandas. Además, la comunidad de código abierto que rodea a Matplotlib ha contribuido al desarrollo de una gran cantidad de recursos, tutoriales y ejemplos que facilitan su aprendizaje y uso.

## Personalización de gráficos

Matplotlib ofrece una amplia gama de opciones para personalizar la apariencia de los gráficos. Esto incluye la capacidad de cambiar los colores de las líneas, los estilos de línea y los marcadores, lo que permite adaptar los gráficos para que se ajusten mejor a las necesidades y preferencias visuales.

### Cambio de colores

Es posible especificar colores usando nombres de colores predefinidos, códigos hexadecimales o valores RGB. Por ejemplo, para cambiar el color de una línea a azul, se puede usar el argumento `color='blue'`.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con color azul
plt.plot(x, y, color='blue')
plt.show()
```

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con color rojo
plt.plot(x, y, color='#Ff0000')
plt.show()
```

### Cambio de estilo de línea

Matplotlib ofrece varios estilos de línea, como sólido ('-'), discontinuo ('--'), puntos ('.'), etc. Se puede especificar el estilo de línea utilizando el argumento `linestyle`.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con línea discontinua
plt.plot(x, y, linestyle='--')
plt.show()
```

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con línea de puntos
plt.plot(x, y, linestyle=':')
plt.show()
```

### Cambio de marcadores

Los marcadores son puntos que se colocan en los puntos de datos. Se puede elegir entre una variedad de formas de marcadores, como círculos ('o'), cuadrados ('s'), triángulos ('^'), etc. Se puede especificar el marcador utilizando el argumento `marker`.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con marcadores de círculo
plt.plot(x, y, marker='o')
plt.show()
```

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar con marcadores de círculo
plt.plot(x, y, marker='+')
plt.show()
```

### Configuración de etiquetado de ejes y título

Una parte crucial de la creación de visualizaciones efectivas es asegurarse de que estén adecuadamente etiquetadas y tituladas. Esto ayuda a una correcta interpretación de la información representada en el gráfico, facilitando la comprensión de este.

En Matplotlib, se puede etiquetar los ejes (ejes x y y) y añadir un título al gráfico utilizando métodos específicos proporcionados por la biblioteca.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico
plt.plot(x, y)

# Etiquetado de ejes
plt.xlabel('Eje X')  # Etiqueta del eje x
plt.ylabel('Eje Y')  # Etiqueta del eje y

# Título del gráfico
plt.title('Gráfico de ejemplo')

# Mostrar el gráfico
plt.show()
```

### Ajuste de los ejes

Una de las características clave de Matplotlib es su capacidad para personalización de gráficos, incluidos los límites de los ejes. Ajustar los límites de los ejes es útil cuando se quiere enfocar un área específica del gráfico o resaltar ciertos detalles.

Para ajustar los límites de los ejes en Matplotlib, se puede utilizar el método `xlim()` para los límites del eje x y `ylim(`) para los límites del eje y.

```python
import numpy as np

# Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y = x

# Crear el gráfico
plt.plot(x, y)
plt.title('Función lineal')
plt.xlabel('X')
plt.ylabel('Y')

# Ajustar los límites de los ejes
#plt.xlim(0, 5)  # Limitar el eje x de 0 a 5
#plt.ylim(0, 5) # Limitar el eje y de -1 a 1

# Mostrar el gráfico
plt.show()
```

Matplotlib permite configurar diferentes escalas en los ejes x e y, como lineal, logarítmica, simétrica, entre otras. Esto puede ser útil cuando los datos tienen un rango amplio o siguen una distribución no lineal. Para poder implementar se usan las funciones `xscale()` y `yscale()`

```python
# Datos de ejemplo
x = [10, 100, 1000, 10000]
y = [1, 2, 3, 4]

# Crear el gráfico con escalas logarítmicas en ambos ejes
plt.plot(x, y)
#plt.xscale('log')  # Escala logarítmica para el eje x
#plt.yscale('linear')  # Escala lineal para el eje y

# Mostrar el gráfico
plt.show()
```

En algunas situaciones, puede ser útil representar dos conjuntos de datos con diferentes unidades en el mismo gráfico, pero con el mismo eje (ya sea el eje x o el eje y) compartido. Para lograr esto, podemos usar los métodos `twinx()` y `twiny()`. El método `twinx()` crea un nuevo conjunto de ejes y los posiciona en la misma ubicación del eje x, mientras que el método `twiny()` hace lo mismo para el eje y.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [20, 25, 30, 35, 40]

fig, ax1 = plt.subplots()

# Graficar el primer conjunto de datos
ax1.plot(x, y1, 'r-')
ax1.set_ylabel('Y1', color='r')

# Crear un segundo conjunto de ejes con el mismo eje x
ax2 = ax1.twinx()
# Graficar el segundo conjunto de datos
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Y2', color='b')

plt.show()
```

### Ajuste de leyendas

Las leyendas son una parte crucial de cualquier gráfico, ya que proporcionan información sobre qué representa cada elemento en el gráfico. Matplotlib ofrece una gran flexibilidad para personalizar las leyendas según las necesidades específicas de cada visualización.

Para ajustar las leyendas en Matplotlib, puedes usar el método `legend()` en el objeto de los ejes (Axes) después de haber añadido los elementos que deseas que aparezcan en la leyenda. Este método acepta varios argumentos que te permiten controlar la apariencia y el comportamiento de la leyenda, como la ubicación, el tamaño del texto y más.

```python
# Generar datos aleatorios
x = np.linspace(0, 10, 50)
y1 = np.random.randint(10, size=50)
y2 = np.random.randint(10, size=50)

# Crear el gráfico de dispersión con dos series de datos
plt.plot(x, y1, label='x_1')
plt.plot(x, y2, label='x_2')

# Ajustar la leyenda
plt.legend(loc='upper right', fontsize='large')

# Mostrar el gráfico
plt.show()
```

### Añadir anotaciones

Una de las características importantes de Matplotlib es su capacidad para añadir anotaciones a los gráficos, lo que permite resaltar puntos clave, proporcionar información adicional o hacer que los gráficos sean más comprensibles. Las anotaciones pueden incluir texto, flechas, formas y otras marcas visuales que ayudan a explicar los datos representados en el gráfico.

Para añadir anotaciones a un gráfico en Matplotlib, podemos utilizar la función `annotate()`. Esta función acepta varios parámetros, incluyendo las coordenadas donde se colocará la anotación, el texto que se mostrará y opcionalmente la apariencia visual de la anotación (como el color y el estilo de línea de la flecha).

```python
# Datos para el gráfico de dispersión
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico de dispersión
plt.scatter(x, y)

# Añadir una anotación en el punto (3, 5)
plt.annotate('Punto interesante', xy=(3, 5), xytext=(3.5, 6),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Etiquetas de los ejes y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión con Anotación')

# Mostrar el gráfico
plt.show()
```

### Cambios de apariencia

Matplotlib ofrece una amplia gama de opciones para personalizar la apariencia general de un gráfico, lo que permite ajustar el tamaño, la resolución y otros aspectos visuales según necesidades específicas.

Para cambiar el tamaño y la resolución de un gráfico, se puede utilizar el parámetro figsize al crear la figura. Este parámetro acepta una tupla de dos valores que representan el ancho y el alto de la figura en pulgadas. Además, se puede especificar la resolución del gráfico utilizando el parámetro dpi (puntos por pulgada).

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear una figura con un tamaño personalizado (8 pulgadas de ancho y 6 pulgadas de alto)
# y una resolución de 100 dpi
plt.figure(figsize=(8, 6), dpi=100)

# Dibujar el gráfico de dispersión
plt.scatter(x, y)

# Etiquetas de los ejes y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión')

# Mostrar el gráfico
plt.show()
```

### Subtramas

Matplotlib permite crear múltiples subtramas dentro de una sola figura, lo que resulta útil cuando se desea comparar diferentes conjuntos de datos o visualizar múltiples gráficos al mismo tiempo. Las subtramas se crean utilizando la función `subplot()` o el método `subplots()` de la clase Figure.

La función `subplot()` toma tres argumentos: el número de filas de subtramas, el número de columnas de subtramas y el índice de la subtrama actual.

```python
# Crear una figura y subtramas
fig, axes = plt.subplots(nrows=2, ncols=2)

# Datos de ejemplo para cada subtrama
x = [1, 2, 3, 4]
y1 = [10, 15, 7, 10]
y2 = [8, 10, 12, 14]
y3 = [15, 13, 11, 9]
y4 = [10, 12, 9, 13]

# Graficar en la primera subtrama
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Subtrama 1')

# Graficar en la segunda subtrama
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Subtrama 2')

# Graficar en la tercera subtrama
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Subtrama 3')

# Graficar en la cuarta subtrama
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Subtrama 4')

# Ajustar diseño
plt.tight_layout()

# Mostrar la figura
plt.show()
```

Matplotlib proporciona varias formas de crear subtramas, una de las cuales es utilizando la función `subplot2grid()`. Esta función permite crear una cuadrícula personalizada de subtramas dentro de una figura, lo que brinda un control preciso sobre su ubicación y tamaño.

El método `subplot2grid()` toma tres argumentos principales: la forma de la cuadrícula de la figura (filas y columnas), la ubicación de la subtrama en la cuadrícula (fila y columna), y opcionalmente, parámetros adicionales como el número de filas y columnas que la subtrama debe abarcar.

```python
# Crear una figura y establecer una cuadrícula de subtramas
fig = plt.figure()

# Utilizar subplot2grid para crear subtramas
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)  # Subtrama que abarca toda la primera fila
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)  # Subtrama que abarca las dos primeras columnas de la segunda fila
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)  # Subtrama que abarca la tercera columna y dos filas
ax4 = plt.subplot2grid((3, 3), (2, 0))  # Subtrama en la tercera fila y primera columna
ax5 = plt.subplot2grid((3, 3), (2, 1))  # Subtrama en la tercera fila y segunda columna

# Personalizar las subtramas
ax1.plot([1, 2, 3], [1, 2, 3])
ax1.set_title('Subtrama 1')
ax2.plot([1, 2, 3], [3, 2, 1])
ax2.set_title('Subtrama 2')
ax3.plot([1, 2, 3], [1, 3, 2])
ax3.set_title('Subtrama 3')
ax4.plot([1, 2, 3], [2, 1, 3])
ax4.set_title('Subtrama 4')
ax5.plot([1, 2, 3], [3, 2, 1])
ax5.set_title('Subtrama 5')

# Ajustar automáticamente el diseño de las subtramas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Integración con Pandas

Matplotlib tiene la capacidad para integrarse fácilmente con otras bibliotecas de análisis de datos en Python, especialmente Pandas. Pandas es una biblioteca popular utilizada para manipular y analizar datos tabulares, principalmente a través de la estructura de datos DataFrame. La integración entre Matplotlib y Pandas permite visualizar rápidamente los datos contenidos en DataFrames de Pandas, lo que facilita la exploración y comprensión de los datos.

Matplotlib puede tomar directamente los datos de un DataFrame de Pandas y usarlos para generar gráficos. Esto se logra pasando las columnas relevantes del DataFrame como argumentos a las funciones de trazado de Matplotlib.

```python
import pandas as pd

# Datos de ejemplo (ventas mensuales)
meses = range(1, 13)
ventas = [20000, 22000, 25000, 28000, 30000, 32000, 35000, 38000, 40000, 42000, 45000, 48000]

# Crear un DataFrame de Pandas
df = pd.DataFrame({'Mes': meses, 'Ventas': ventas})

# Crear el gráfico de línea
plt.plot(df['Mes'], df['Ventas'], marker='o')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

## Tipos de gráficos

### Gráficos de línea

Los gráficos de línea son una de las formas más comunes de visualización de datos y son ideales para mostrar tendencias a lo largo del tiempo o en una secuencia ordenada. En Matplotlib, los gráficos de línea se crean utilizando la función `plot()`.

```python
# Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico de línea
plt.plot(x, y)

# Añadir etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Línea')

# Mostrar el gráfico
plt.show()
```

### Gráficos de dispersión

Los gráficos de dispersión son una forma eficaz de visualizar la relación entre dos variables numéricas. Cada punto en el gráfico representa una observación individual y su posición en los ejes x e y corresponde a los valores de estas variables. Este tipo de gráficos es útil para identificar patrones, tendencias, y posibles correlaciones entre las variables. En Matplotlib, los gráficos de dispersión se crean utilizando la función `scatter()`.

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico de dispersión
plt.scatter(x, y, color='blue', marker='o', label='Datos')

# Añadir etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

### Histogramas

Los histogramas son una herramienta fundamental en el análisis de datos para visualizar la distribución de una variable numérica. Permiten ver la frecuencia con la que ocurren diferentes valores dentro de un conjunto de datos y proporcionan información sobre la forma y la dispersión de la distribución. En Matplotlib, los histogramas se crean utilizando la función `hist()`.

```python
import numpy as np

# Generar datos aleatorios
datos = np.random.randn(1000)

# Crear el histograma
plt.hist(datos, bins=20, edgecolor='black')

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos Aleatorios')

# Mostrar el histograma
plt.show()
```

### Gráficos de barras

La gráfica de barras es una de las visualizaciones más comunes y útiles para representar datos discretos o categorizados. Se utiliza principalmente para mostrar la comparación entre diferentes grupos o categorías. Cada barra representa una categoría particular, y la altura de la barra indica la magnitud de la variable que se está representando. En Matplotlib, los gráficos de barrras se crean utilizando la función `bar()`.

```python
# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 33]

# Crear la gráfica de barras
plt.bar(categorias, valores)

# Añadir etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')

# Mostrar la gráfica
plt.show()
```

### Gráficos de pastel

Los gráficos de pastel, también conocidos como gráficos circulares, son una forma popular de representar datos categóricos en forma de un círculo dividido en segmentos. Cada segmento del pastel representa una categoría de datos y el tamaño del segmento es proporcional a la cantidad de datos que representa en relación con el conjunto completo de datos. Los gráficos de pastel son útiles para mostrar la composición relativa de diferentes categorías y facilitan la comparación visual entre ellas. En Matplolib, los gráficos de pastel se crean utilizando la función `pie()`.

```python
# Datos
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Porcentajes de cada categoría
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Colores para cada categoría
explode = (0, 0.1, 0, 0)  # Resaltar una categoría específica

# Crear el gráfico de pastel
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Ajustar aspecto
plt.axis('equal')  # Proporciones iguales para asegurar que el pastel sea circular

# Añadir título
plt.title('Gráfico de Pastel')

# Mostrar el gráfico
plt.show()
```

### Gráficos de caja

Los gráficos de caja, también conocidos como diagramas de caja o boxplots, son una forma de representar gráficamente la distribución de un conjunto de datos numéricos a través de sus cuartiles. Son útiles para visualizar la dispersión y la simetría de los datos, así como para identificar valores atípicos (outliers). En Matplotlib, los gráficos de caja se crean utilizando la función `boxplot()`.

```python
# Generar datos aleatorios
np.random.seed(10)
datos = np.random.normal(loc=0, scale=1, size=100)

# Crear el gráfico de caja
plt.boxplot(datos)

# Añadir etiquetas y título
plt.xlabel('Datos')
plt.ylabel('Valores')
plt.title('Gráfico de Caja')

# Mostrar el gráfico
plt.show()
```

### Gráficos de contorno

Los gráficos de contorno, también conocidos como mapas de contorno, son útiles para visualizar datos tridimensionales en dos dimensiones mediante líneas de nivel. Son especialmente útiles para representar funciones de dos variables, donde las líneas de contorno muestran los puntos donde la función tiene el mismo valor.

Estos gráficos son comunes en campos como la meteorología, la topografía, la física y la ingeniería, donde se utilizan para representar campos escalares, como la temperatura, la presión o la elevación, en un plano bidimensional.
En Matplolib, los gráficos de contorno se crean utilizando la función `contour()`.

```python
# Generar datos para el contorno
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear el gráfico de contorno
plt.contour(X, Y, Z, levels=20)  # levels controla la cantidad de líneas de contorno
plt.colorbar()  # Añadir una barra de color para visualizar los valores
plt.title('Gráfico de Contorno de $\sin(\sqrt{X^2 + Y^2})$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Gráficos de superficie

Los gráficos de superficie son una forma de visualización tridimensional que muestra la relación entre tres variables. Se utilizan comúnmente para representar datos que varían continuamente sobre una superficie bidimensional. Estos gráficos son útiles para visualizar funciones matemáticas tridimensionales, como funciones de densidad, funciones de distribución conjunta, campos vectoriales, entre otros.

En Matplotlib, los gráficos de superficie se pueden crear utilizando la función `plot_surface()` del módulo `mpl_toolkits.mplot3d`, que proporciona herramientas para crear gráficos tridimensionales.

```python
from mpl_toolkits.mplot3d import Axes3D

# Datos de ejemplo: coordenadas x, y, z
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear el gráfico de superficie
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Añadir etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico de Superficie')

# Mostrar el gráfico
plt.show()
```

### Gráficos tridimensionales


Matplotlib ofrece la capacidad de crear visualizaciones en 3D, lo que permite representar datos tridimensionales de manera efectiva. Estos gráficos son útiles cuando se trabaja con conjuntos de datos que tienen una dimensión adicional, lo que permite explorar la relación entre tres variables de forma visual.

```python
# Datos de ejemplo
theta = np.linspace(0, 2*np.pi, 100)
x = np.sin(theta)
y = np.cos(theta)
z = theta

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
line, = ax.plot(x, y, z)

plt.show()
```

### Mapas de calor

Los mapas de calor son una forma efectiva de visualizar datos bidimensionales mediante la asignación de colores a valores en una matriz. Son útiles para representar la densidad de puntos en un área y para resaltar patrones o variaciones en los datos.

En Matplotlib, para crear una mapa de calor se utiliza la función `imshow()`.

```python
# Crear el mapa de calor
import numpy as np
import matplotlib.pyplot as plt

datos = np.ones((2, 2))
#datos[1,:] = 1
plt.imshow(datos, cmap='hot', interpolation='nearest')

# Agregar barra de color
plt.colorbar()

# Mostrar el mapa de calor
plt.show()
```

### Diagramas de Violín

Los diagramas de violín son una representación gráfica que combina las ventajas de un diagrama de caja y bigotes con un gráfico de densidad de Kernel. Muestran la distribución de un conjunto de datos numéricos a través de un núcleo simétrico alrededor de su mediana. Esto proporciona una representación más completa de la distribución de los datos, incluyendo información sobre la densidad y la forma de la distribución, además de los estadísticos de resumen tradicionales. En Matplotlib, para crear un diagrama de violín se utiliza la función `violinplot()` en conjunto con la biblioteca Seaborn.

```python
import seaborn as sns  # Biblioteca Seaborn

# Datos de ejemplo
data = sns.load_dataset('iris')

# Crear un diagrama de violín
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=data)
plt.title('Diagrama de Violín de Longitud del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del Sépalo')
plt.show()
```
